```python
# simple calculator
import tkinter as tk
from tkinter import *

#create windows 
win = tk.Tk()
win.geometry("300x400")
win.resizable(0, 0)
# set title for an application
win.title('Simple Calculator')

# btn click set inpute expression
def btn_click(item):
    global expression
    expression = expression + str(item)
    e_var.set(expression)
    
# clear btn
def bt_clear():
    global expression 
    expression = "" 
    e_var.set("")

#  btn qual opration on    
def btn_equl():
    global expression
    result = str(eval(expression)) # 'eval':This function is used to evaluates the string expression directly
    e_var.set(result)
    expression = ""
    
expression = ""

# input frame
i_frame = tk.Frame(win,width=300,height=50,highlightbackground="black",highlightcolor="black",highlightthickness=2)
i_frame.pack(side=TOP)

# i_entry variablr
e_var = tk.StringVar()

# entry in frame
i_entry = tk.Entry(i_frame,font=('arial', 18, 'bold'),textvariable=e_var,width=40,justify=RIGHT)
i_entry.grid(row=0, column=0)
i_entry.pack(ipady=10)

# button frame
b_frame = Frame(win, width=300, height=349.5, bg="grey",highlightthickness=2)
b_frame.pack()
# button
# row 0
seven = tk.Button(b_frame, text = "7", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(7)).grid(row = 0, column = 0, padx = 1, pady = 1)
eight = tk.Button(b_frame, text = "8", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(8)).grid(row = 0, column = 1, padx = 1, pady = 1)
nine = tk.Button(b_frame, text = "9", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(7)).grid(row = 0, column = 2, padx = 1, pady = 1)
dived = tk.Button(b_frame, text = "/", fg = "black", width = 9, height = 3, bd = 0, bg = "#eee", cursor = "hand2", command = lambda: btn_click("/")).grid(row = 0, column = 3, padx = 1, pady = 1)
# row 1
four = tk.Button(b_frame, text = "4", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(4)).grid(row = 1, column = 0, padx = 1, pady = 1)
five = tk.Button(b_frame, text = "5", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(5)).grid(row = 1, column = 1, padx = 1, pady = 1)
six = tk.Button(b_frame, text = "6", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(6)).grid(row = 1, column = 2, padx = 1, pady = 1)
mult = tk.Button(b_frame, text = "*", fg = "black", width = 9, height = 3, bd = 0, bg = "#eee", cursor = "hand2", command = lambda: btn_click("*")).grid(row = 1, column = 3, padx = 1, pady = 1)

# row 2
one = tk.Button(b_frame, text = "1", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(1)).grid(row = 2, column = 0, padx = 1, pady = 1)
two = tk.Button(b_frame, text = "2", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(2)).grid(row = 2, column = 1, padx = 1, pady = 1)
three = tk.Button(b_frame, text = "3", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(3)).grid(row = 2, column = 2, padx = 1, pady = 1)
sub = tk.Button(b_frame, text = "-", fg = "black", width = 9, height = 3, bd = 1, bg = "#eee", cursor = "hand2", command = lambda: btn_click("-")).grid(row = 2, column = 3, padx = 1, pady = 1)

# row 3
point = tk.Button(b_frame, text = ".", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(".")).grid(row = 3, column = 0, padx = 1, pady = 1)
zero = tk.Button(b_frame, text = "0", fg = "blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda: btn_click(0)).grid(row = 3, column = 1, padx = 1, pady = 1)
equl = tk.Button(b_frame, text = "=", fg = "white",bg="blue", width = 9, height = 3, bd = 1,cursor = "hand2", command = lambda:btn_equl()).grid(row = 3, column = 2, padx = 1, pady = 1)
plus = tk.Button(b_frame, text = "+", fg = "black", width = 9, height = 3, bd = 1, bg = "#eee", cursor = "hand2", command = lambda: btn_click("+")).grid(row = 3, column = 3, padx = 1, pady = 1)

# row 4
clear = tk.Button(b_frame, text = "C", fg = "black", width = 32, height = 3, bd = 1, bg = "#eee", cursor = "hand2", command = lambda: bt_clear()).grid(row = 4, column = 0, columnspan = 3, padx = 1, pady = 1)
 

win.mainloop()
```
